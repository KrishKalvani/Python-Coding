```python
'''
QUESTION 1 - (1) - Two Sum
Given an array of integers nums and an integer target,
return indices of the two numbers such that they add up to target.


In the function, we create an empty dictionary that stores the numbers in nums array as KEYS and their index values as VALUES.
now it is empty. but it will look like, nums=[0:1, 1:2]

in the for loop, in each iteration, enumerate will give us the actual number in the array along with it's index value.
1st iteration: index=0, num=1
using this, we calculate the difference between target (5) and the number (1), and we get 4, 
but it is not there in the num_to_index dictionary (cause its empty), 
so the if block is skipped and we do this: num_to_index[num] = i
We add the current number (1) with its index (0) to the dictionary
num_to_index = {0:1}

2nd iteration, same process.
i=1, num=2
5-2 =3
not there, put it in the dictionary
{0:1, 1:2}

3rd iteration
i-2, num=3
5-3=2
2 is there in the dictionary, so we return the index of 2 from the dict and index of [i] - current number(3)
=[1,2]

            
    
'''

def two_sum(nums, target): #function with parameters - nums array, and target value
    num_to_index = {} # create an empty dictionary to store keys(nums) and values(indices of the num in nums)
    
    for i, num in enumerate(nums): #iterate through each number with their respective index values.
        differenceVal = target - num #find the difference between the target and the current nummber
        
        if differenceVal in num_to_index: #if difference value is there in the dictionary
            return [num_to_index[differenceVal], i] #then return the index of that difference value, and index of current number.
        
        num_to_index[num] = i #if difference value is not there, then add the current num in the dictionary with its index value
        
nums=[1,2,3,4] #filling our inputs
target=5
result=two_sum(nums,target)
print(result)
```

```python
'''QUESTION 3 - (20) - Valid Parentheses

Given a string s containing just the characters '(', ')', '{', '}', '[' and ']', determine if the input string is valid.

Validity conditions:
- Open brackets must be closed by the same type of bracket
- Open brackets must be closed in the correct order
- Every closing bracket must have a corresponding open bracket of the same type

I have used a stack data structure to keep track of open brackets that need closing.
When iterating through string s, we will encounter an open bracket (, or [, or {, then we add them to 
the data structure.
When encountering a closing bracket, if it disobeys any of the validity conditions, then return false.

LOGIC:
create a bracket map to to assign closing brackets to the correct opening brackets
create an empty stack
iterate through the input e.g., '()'
1st iteration for first char '(', it is not in bracket_map (its an open bracket),
append '(' to the stack. stack=['(']
2nd iteration for ')' char
it is in the bracket_map (closing bracket),
pop the last element ('(') in the stack. Now stack=[]
top_element variable will store the popped element. If top_element matches the
current ')', then return true. If it doesnt match, return false.
If the stack is already empty because the user enters invalid input at first,
then assign '#' dummy value to top_element and use that for match checking,
it will give false cause the input was invalid. This is done to handle errors.
'''

def is_valid(s):
    bracket_map = {')' : '(', '}' : '{', ']' : '['} #map each closing bracket with their opening brackets.
    
    stack=[] #stores open brackets
    
    for char in s: #iterate through each char in the input s
        if char in bracket_map: #if char is a closing bracket
            top_element=stack.pop() if stack else '#'
            #remove the last element in the stack. Store it in top_element 
            #If stack is already empty (if invalid input)
            #assign dummy value '#' to top_element to prevent errors (will give false)
            
            if bracket_map[char] != top_element: #if the top_element value matches 
                #with the bracket_map, then true
                #otherwise, return false
                return False
        else: #if char is an opening bracket
            stack.append(char) #then push it into the stack
    return not stack

print(is_valid("()"))
```

```python
'''QUESTION 7 - (66) - Plus one.

You are given a large integer represented as an integer array digits, where each digits[i] is the ith digit of the integer. 
The digits are ordered from most significant to least significant in left-to-right order. The large integer does not contain any leading 0's.
Increment the large integer by one and return the resulting array of digits.'''

'''
digits array is given in the input for e.g., [1,2,4]
the loop iterates/traverses the digits array from the last digit (LSB)
first check if the current digit is less than 9, if so, add 1 and return the result.
if the current digit is 9, set it to 0 and continue to the next left digit and
carry over the 1. By chance, if all the digits were 9, they will eventually become 0.
So, in this case, we prepend 1 as the MSB at the front.
'''

def plus_one(digits):
    for i in range(len(digits) -1, -1, -1): #iterating from right to left
    
        if digits[i] < 9: #if the current number is less than 9
            digits[i] += 1 #then add 1
            return digits #return the result
    
        digits[i] = 0 #if the current digit is 9, set it to 0 (carry over 1)
    return [1] + digits #use the carried over 1, and add it as the MSB at the front.

print(plus_one([1,2,3]))
print(plus_one([4,3,2,1]))
print(plus_one([9,9,9]))

    
```

```python
'''QUESTION 8 - (69) - Sqrt(x)

Given a non-negative integer x, return the square root of x rounded down to the nearest integer. 
The returned integer should be non-negative as well. You must not use any built-in exponent function or operator.'''

'''
here we have to use binary search. The square root of x would lie between 0 and x.
if x is very small like 0 or 1, then we directly return x as the result.
With binary search, we have to calculate the mid value i.e., low + high/2
then check if squaring the mid value gives the exact match as x. if it does, 
then mid is the square root.
if mid*mid < x, we move the search to the right half because the square root must be greater than mid
if mid*mid > x, move to left half because sqrt must be smaller than mid.

'''

def my_sqrt(x):
    if x < 2: #if input is less than 2
        return x #return the input itself
    low = 0 #low value is 0
    high = x #high value is the input, e.g., x=8
    result = 0
    
    while low <= high: #iterate through the range from 0 to x (8)
        mid = (low + high)//2 #calculate mid e.g., 0+8/2 = 4
        if mid * mid == x: # 16!== 8 , skip this block
            return mid
        elif mid * mid < x: #16 is greater than 8, not lesser, skip this block.
            result = mid
            low = mid + 1
        else: #16 is greater than 8
            high = mid - 1 #4-1 = 3 , now high = 3 (first iteration)
    return result
#for the 2nd iteration, the new range will be 0,3 - low = 2,
#3rd iteration, range will be 2,3, low = mid + 1, so 2+1=3
#4th iteration, range is 3,3, mid=3, 9>8, so high = mid -1 (3-1) = 2
# now low is 3, high is 2, low>high, loop ends, return the last valid mid (2) as the result.

print(my_sqrt(8))
```

```python
'''
QUESTION 10 - (70) - Climbing Stairs.

You are climbing a staircase. It takes n steps to reach the top.
Each time you can either climb 1 or 2 steps. In how many distinct ways can you climb to the top?
reference: https://github.com/doocs/leetcode/blob/main/solution/0000-0099/0070.Climbing%20Stairs/README_EN.md
f(n) = f(n-1) + f(n-2)
'''

def climb_stairs(n: int) -> int: #expect an int input and return an int input.
    
    if n == 1: #if total steps is 1
        return 1 #return 1 because there's only 1 way
    if n == 2: #if total steps is 2
        return 2 #return 2 because there are only 2 ways

    #initializing variables for f(n-1) and f(n-2)
    prev_step1 = 2 #this is f(2), it represents f(n-1) for later
    prev_step2 = 1 #this is f(1), it represents f(n-2) for later

    #variable to store the result for the current n
    current_step = 0

    #calculate for n = 3 to n = 5
    for i in range(3, n + 1):
        
        current_step = prev_step1 + prev_step2  #current result (f(n)) = f(n-1) + f(n-2)
        prev_step2 = prev_step1  #update f(n-2) to be the previous f(n-1)
        prev_step1 = current_step  #update f(n-1) to be the current result

    return current_step

print(climb_stairs(5))
```

```python
'''
QUESTION 16 - (3024) - Type of Triangle

You are given a 0-indexed integer array nums of size 3 which can form the sides of a triangle.
A triangle is called equilateral if it has all sides of equal length.
A triangle is called isosceles if it has exactly two sides of equal length.
A triangle is called scalene if all its sides are of different lengths.
Return a string representing the type of triangle that can be formed or "none" if it cannot form a triangle. 
'''

'''
here, first we should use the triangle inequality theorem, where the sum of the lengths
of any 2 sides must be greater than the length of the 3rd side. Without this condition,
the sides cannot form a triangle

we start by sorting the nums array so that nums[0] is the smallest side, nums[1]
is the middle side, and nums[2] is the largest side. This makes it easier to check the triangle
inequality condition.

'''

def triangle_classifier(nums):
    
    if len(nums) != 3: #input validity check
        return "Invalid input, must have exactly 3 sides."
    
    nums.sort() #sort the array, ascending order.
    
    if nums[0] + nums[1] <= nums[2]: #if the sum of the 2 sides is lesser than the 3rd side,
                                     #then, its not a valid triangle
        return "None"
    
    if nums[0] == nums[1] == nums[2]: #if all sides are equal
        return "equilateral"
    
    #if exactly 2 sides are equal
    elif nums[0] == nums[1] or nums[1] == nums[2] or nums[0] == nums[2]:
        return "isosceles"
    else:
        return "scalene"

    
#input testing
print(triangle_classifier([5,5,5]))
print(triangle_classifier([5,5,8]))
print(triangle_classifier([4,5,6]))
print(triangle_classifier([1,1,5]))
print(triangle_classifier([5,5,5,5]))
print(triangle_classifier([5,5]))
```

```python
'''
QUESTION 31 - (567)  Permutation in String

Given two strings s1 and s2, return true if s2 contains a 
permutation of s1, or false otherwise.
In other words, return true if one of s1's permutations is the substring of s2.


LOGIC:
s1 = "ab"
s2 = "eidbaooo"

we just have to check if s2 contains the same letters of s1 even if it is rearranged, but not away from each other, for example, "ab" or "ba" are
valid permuations. If s2 was eidbooao, then thats not a valid permutation.
if s1 is longer than s2 , then thats impossible, we can make that as our base case.

we will import and use the Counter package from the collections module to count the frequency
of letters for s1
s1_count = {'a': 1. 'b': 1}

we will use the sliding window technique by initializing the first window_count variable.
This variable will only store the first 2 characters of the s2 and if neither 'a' or'b' are
consecutively matching, it will slide to the next letters until it finds a match.
window_count = {'e' : 1, 'i' : 1}

s2 = "eidbaooo"
so in the first window/first iteration, the letters e and i didnt match a or b,
so we proceed the rest of the string, we add 'd' and remove 'e'
window_count = {'i': 1, 'd': 1}
again, no match, so we do the same thing:

2nd iteration: add 'b', remove 'i' -> window_count = {'d': 1, 'b': 1}, here 'b' is there, but not 'a'
3rd iteration: add 'a', remove 'd' -> window_count = {'b': 1, 'a': 1}

now match is found with s1_count, so return true.


we use the for loop to slide the window across s2, if there is a match of s1 letters,
return True.

'''

from collections import Counter

def check_permutation(s1, s2):
    #if s1 is longer than s2, its impossible
    if len(s1) > len(s2):
        return False

    #frequency count for s1
    s1_count = Counter(s1)

    #initialize a window in s2 - window_count = {'e' : 1, 'i' : 1}
    window_count = Counter(s2[:len(s1)])

    #slide the window across s2 (loop through s2 string)
    for i in range(len(s1), len(s2)):
        #check if the current window matches
        if window_count == s1_count:
            return True

        #slide the window: add the new character
        window_count[s2[i]] += 1

        #remove the old character from window_count
        start_char = s2[i - len(s1)] #identify the start character to be removed.
        window_count[start_char] -= 1 #remove it, now the 2nd item comes to the first position, so there's place for the newly added item.
        if window_count[start_char] == 0: #if the frequency becomes 0, then delete it.
            del window_count[start_char]

    #final check for the last window
    return window_count == s1_count

s1="ab"
s2="eidbaooo"
print(check_permutation(s1,s2))

```

```python
'''
QUESTION 34 - (394) Decode String

Given an encoded string, return its decoded string.

The encoding rule is: k[encoded_string], where the encoded_string inside the square brackets is being repeated exactly k times. 
Note that k is guaranteed to be a positive integer.

Input: s = "3[a]2[bc]"
Output: "aaabcbc"

LOGIC:
using the example input above,
we will create an empty stack to append the input characters into it and pop the elements
to decide how many times to repeat it depended on the digit before the opening of the square brackets.

current_num = 0
current_string = ""


in a for loop, we will iterate through each input character in the string
= "3[a]2[bc]", then first we will check if the current string character is a digit,
if so, convert it into a digit and store the number in current_num variable.

if the next character is a '[', then append the current_num = 3 into the stack
and as well as the current_string = "" in the stack.
stack = ("", 3)
reset current_String = "", and current_num  = 0

iteration over a normal letter e.g., 'a', we simply add 'a' to the current_string,
now current_string = "a"

now when the loop encounters the ']' bracket, this is where we take the number 3 from the stack
and tell the current string to repeat the prev_string "" + the current string "a" x 3 times.
= "aaa"

current_string = prev_string + current_string * repeat_times

same process for 2[bc]

iteration over '2':
current_num = 2

iteration over '[':
push ("aaa", 2) onto the stack.

reset current_string = "" and current_num = 0
iteration over 'b' and 'c':

add 'b' and 'c' to current_string: current_string = "bc"
tteration over ']':

pop the stack: prev_string = "aaa", repeat_times = 2.
current_string = "aaa" + "bc" * 2 = "aaabcbc"

return current_string which is decoded.
'''

def decodeString(s):
    stack = [] #stack to store characters, numbers, and results
    current_num = 0
    current_string = ""

    for char in s: #iterate through all characters in the input
        if char.isdigit(): #if string char is a digit,
            #build the current number
            current_num = current_num * 10 + int(char)
        elif char == '[':
            #push the current string and current number to the stack
            stack.append((current_string, current_num))
            #reset for new segment (next iteration)
            current_string = ""
            current_num = 0
        elif char == ']':
            #pop (extract from) the stack and repeat the current string k times
            prev_string, repeat_times = stack.pop()
            current_string = prev_string + current_string * repeat_times
        else:
            #append characters to the current string if its simply just a regular char
            current_string += char

    return current_string #return the decoded string.

s="3[a]2[bc]"
print(decodeString(s))
```

```python
'''
QUESTION 36 - (767) - Reorganize string

Given a string s, rearrange the characters of s so that any two adjacent characters are not the same.
Return any possible rearrangement of s or return "" if not possible.


LOGIC:

we have to return the rearranged version string s such that the same consecutive characters
are not next to each other.

for e.g.,
s = "aab"
we want to rearrange "aab" to a different manner such that it becomes "aba"

We use the counter from collections module,
we count the frequenices of 'a' and 'b'. This is stored as a dictionary.
char_counts = {'a':2, 'b':1}
now we want to sort the characters based on how frequently they appear in descening order i.e.,
with the most frequent character first.
we use char_counts.items() to convert the Counter dictionary into a list of tuples:
[('a',2), ('b',1)] and each tuple contains a character and its frequency.

we will use key=lambda x: -x[1], to sort the list using a negative frequency
the -ve sign is needed cause we want the most frequent characters first.
(python sorts in ascending order by default).

the sorted funtion applied this sorting rule to the list of tuples.

now we check if the most frequent charcter can be placed without adjacent repeats.
if sorted_chars[0][1] > (len(s) + 1) // 2:
return ""


this means that if the most frequent character appears more than half the 
length of the string (rounded up), then it is impossible to arrange the characters
without adjacent repeats. So we return an empty string as the output.

for the "aab" example
(if '2' > 3+1//2) = if '2' > 2, condition will be false, so we can arrange it.

now we create a result array to hold the reaaranged characters.
result = [""] * len(s)
for s = "aab", it becomes: result = ["", "", ""]

we will create a tracker i.e., idx = 0, to track the position in the result array
where the next character will be placed.

so for e.g., 'a' will need to be placed twice, that was the highest frequency in
the tuple, so by default, if we see the highest frequency, we will increment
idx by 2 to avoid the adjacent placement next to the 0th position in the results
array.




'''


from collections import Counter

def rearrangeString(s):
    #count character frequencies
    char_counts = Counter(s)
    
    #sort characters by frequency (highest first)
    sorted_chars = sorted(char_counts.items(), key=lambda x: -x[1])
    
    #check if rearrangement is possible
    if sorted_chars[0][1] > (len(s) + 1) // 2:
        return ""
    
    #arrange characters in a result array
    result = [""] * len(s)
    idx = 0 #first character is placed at 0th position of results array.
    # later we increment idx by 2 to avoid adjacent placements.
    
    #distributing characters
    for char, count in sorted_chars: #loop through the tuple a:2 b:1
        for _ in range(count): #places current character 'count' times in the result array, so 2 times for 'a'.
            
            result[idx] = char
            idx += 2
            if idx >= len(s): #if idx goes beyond the array size
                idx = 1 #reset it to 1
    
    return "".join(result) #joins the list result into a single string and returns it.


s = "aab"
print(rearrangeString(s))
```

```python
'''
QUESTION 42 - (68) - Text Justification


Given an array of strings words and a width maxWidth, format the text such that each line has exactly maxWidth characters and is fully (left and right) justified.
You should pack your words in a greedy approach; that is, pack as many words as you can in each line. Pad extra spaces ' ' when necessary so that each line has exactly maxWidth characters.
Extra spaces between words should be distributed as evenly as possible. If the number of spaces on a line does not divide evenly between words, the empty slots on the left will be assigned more spaces than the slots on the right.
For the last line of text, it should be left-justified, and no extra space is inserted between words.

LOGIC:
For example,
words = ["This", "is", "an", "example", "of", "text", "justification."]
maxWidth = 16

we will build our resulted list by building the current_line list
we will keep adding each word per iteration,
if adding the word occupies more than 16 characters in the line, then we
calculate the spaces required to distribute evenly and go to the next line.

calculation example:

The total spaces we need to distribute are spaces_needed. The number of gaps between words is len(current_line) - 1.
spaces_per_gap = spaces_needed // (len(current_line) - 1)

the spaces might not divide evenly, we calculate how many "extra spaces" we need to distribute. 
The remainder from the division (spaces_needed % (len(current_line) - 1)) tells us how many extra
spaces to distribute to the leftmost gaps.

example rough calculation:

current_line = ["This", "is", "an"]
maxWidth = 16
current_length = 6 (so we need 16 - 6 = 10 spaces)
spaces_per_gap = 10 // 2 = 5  so there are two gaps between three words
extra_spaces = 10 % 2 = 0  and no extra spaces, since 10 divides evenly

to build the final line: 
start with "This".
add 5 spaces between "This" and "is".
add 5 spaces between "is" and "an".


["This     is     an"]

The process for the remaining word goes on like this by  
going through the same process of checking if we can add the word or if we need to justify the line.

The last line is treated differently from all the other lines because, for the last line:

left justification is used meaning its aligned to the left margin
and any remaining space, after all the words have been placed, is added at the end of the line, after the last word.

no extra spaces are inserted between words.




'''

def fullJustify(words, maxWidth):
    
    result = [] #stores the final adjusted string
    current_line = [] #stores how many words are currently processed. This builds up through the iterations.
    current_length = 0 #total current length of the building string

    for word in words: #iterate through each word in the input
        
        #if adding this word would exceed maxWidth, justify the current line
        if current_length + len(word) + len(current_line) > maxWidth:
            #find the amount of spaces needed to reach the target maxWidth
            spaces_needed = maxWidth - current_length
            if len(current_line) == 1: #if input is just one word, then append all spaces to it
                #all the spaces will be added after that word to make the line exactly maxWidth characters wide.
                result.append(current_line[0] + ' ' * spaces_needed)
            

#If the line contains more than one word, we need to distribute spaces evenly 
#between the words to ensure that the total length is exactly maxWidth. 
#If the spaces do not divide evenly, the "extra spaces" will be added starting from the leftmost gap.


            else:
                #distribute spaces evenly between words
                spaces_per_gap = spaces_needed // (len(current_line) - 1) #this gives us the minimum number of spaces to add between each pair of words.
                extra_spaces = spaces_needed % (len(current_line) - 1) #tells us how many extra spaces to distribute to the leftmost gaps.
                
                #construct the final line
                line = current_line[0] #start with the first word (no spaces before it)
                for i in range(1, len(current_line)): #iterate through each word in the current_line which has the built words
                    #add the calculated number of spaces.
                    
                    # add spaces_per_gap spaces, plus one additional space 
                    #if this is one of the "extra spaces" that need to be distributed.
                    spaces = ' ' * (spaces_per_gap + (1 if i <= extra_spaces else 0))
                    line += spaces + current_line[i]
                result.append(line)
            
            #reset for the next line in the next iteration.
            current_line = []
            current_length = 0
        
        #if the word added didn't exceed maxWidth, then add the word to the current_line
        current_line.append(word)
        current_length += len(word) #update the new current_length
    
    # Handle the last line, which is left-justified
    last_line = ' '.join(current_line) #normally join the words from current_line together with a single space between them
    
    #alculate the remaining spaces needed to reach the maxWidth.
    last_line += ' ' * (maxWidth - len(last_line))
    result.append(last_line) #add the remaining string of spaces to the end of last_line to ensure it is exactly maxWidth characters long.
    
    return result #give the output


words = ["This", "is", "an", "example", "of", "text", "justification."]
maxWidth = 16

print(fullJustify(words,maxWidth))
```

```python
'''
QUESTION 45 - (1406) - Stone Game III


Alice and Bob continue their games with piles of stones. There are several stones arranged in a row, and each stone has an associated value which is an integer given in the array stoneValue.
Alice and Bob take turns, with Alice starting first. On each player's turn, that player can take 1, 2, or 3 stones from the first remaining stones in the row.
The score of each player is the sum of the values of the stones taken. The score of each player is 0 initially.
The objective of the game is to end with the highest score, and the winner is the player with the highest score and there could be a tie. The game continues until all the stones have been taken.
Assume Alice and Bob play optimally.
Return "Alice" if Alice will win, "Bob" if Bob will win, or "Tie" if they will end the game with the same score.

LOGIC:

The game involves Alice and Bob picking stones optimally. To decide the winner, 
we calculate the score difference for all possible scenarios.

we will use the bottom-up approach i.e., we calculate the best score difference for each position from end to start, because
to decide the best move for Alice at a position, we need to know the results of the remaining stones.

reference: https://walkccc.me/LeetCode/problems/1406/#__tabbed_2_3

example:

stoneValue = [1,2,3,7]
n = 4
dynamic programmning (dp) or best_score_diff = [-inf, -inf, -inf, -inf, 0] we start with -inf for each position, then progress it from there.

backward calculation - 

i = 3 (last stone)
alice chooses the number of stones that maximizes the score difference.
score difference = total stones alice takes - best outcome for bob in the remaining stones(dp[i+k])
so if taking 1 stone results in a smaller score difference than taking 2 or 3 stones, she won't choose it.

alice takes 1 stone, she chooses 7 because it will have the best score difference.
dp[3] = max(-inf, 7 - 0) = 7

i = 2 (2nd last stone)
alice takes one stone 3 = dp[2] = max(-inf, 3 - 7) = -4
alice takes two stone 3+7 = dp[2] = max(-4, 10 - 0) = 10

i = 1 (3rd last stone)
alice takes one stone 2 = dp[1] = max(-inf, 2 - 10) = -8
alice takes two stones (2 + 3) = dp[1] = max(-8, 5 - 7) = -2

it goes on, and if the dp[0] is negative, bob will win, otherwise alice will win,
if the value is 0 then its a tie.




'''




# class Solution: -> here use of class is optional
def stoneGame3(stoneValue):
        n = len(stoneValue)
        
        #-inf is the smallest possible value, and we use it to calculate the maximum value
        #starting with -inf ensures that any real score difference will replace it.
        dp = [float('-inf')] * (n + 1)  #initialize dp array with -infinity.
                                                     #this array will store the best score difference from each position in the game.
            
        dp[n] = 0  #when no stones are left, difference is 0
        
        #process the game backwards
        for i in range(n - 1, -1, -1): #starting from the last stone and go to the first.
            total = 0
            for k in range(1, 4):  #alice can take 1, 2, or 3 stones
                if i + k <= n: #ensure alice doesn't go beyond the last stone
                    total += stoneValue[i + k - 1] #add the stone values that alice picks
                    
                    #update dp[i] with the best score difference
                    dp[i] = max(dp[i], total - dp[i + k])
    
        #determine the winner
        if dp[0] > 0: #alice's score is higher
            return "Alice"
        elif dp[0] < 0: #bob's score is higher
            return "Bob"
        else: #scores are the same.
            return "Tie"

stoneValue = [1,2,3,7]
# solution = Solution() -> here use of class is optional
# result = solution.stoneGame3(stoneValue)        
print(stoneGame3(stoneValue))
```

```python
'''
QUESTION 47 - (2344) - Minimum Deletions to Make Array Divisible

You are given two positive integer arrays nums and numsDivide. You can delete any number of elements from nums.
Return the minimum number of deletions such that the smallest element in nums divides all the elements of numsDivide.
If this is not possible, return -1.
Note that an integer x divides y if y % x == 0.


LOGIC:

we basically have to find the smallest number and check if that number is divisble with all
the numbers from the numsDivide array, if that number is not divisible with not all of the numbers in the numsDivide array,
then we delete that number (along with its duplicated occurences).

example of calculating the gcd using the a and b value pairs from the numsDivide array

nums = [2, 3, 2, 4, 3]
numsDivide = [9, 6, 9, 3, 15]

input: a = 9, b = 6
iteration 1: a = 6, b = 9 % 6 = 3
iteration 2: a = 3, b = 6 % 3 = 0
output: 3 (gcd of 9 and 6).

next calculate the gcd fo the list of numbers
nums = [9,6,3]
initial -> result = 9
iteration 1: result = gcd(9,6) = 3
iteration 2: result = gcd(3,3) = 3
o/p = 3.




'''


class Solution:
    def gcd(self, a, b): #calculates the greatest common divisor of two numbers i.e., a and b.
        while b != 0: #keep finding the gcd until b = 0
            a = b #swap values
            b = a%b #replace b with the remainder of a%b
        return a
    
    def find_gcd(self, nums): #now calculate the gcd for the list of numbers
        #compute gcd of a list of numbers
        result = nums[0] #start with the first number
        for num in nums[1:]: #loop through the rest of the numbers
            result = self.gcd(result, num) #update the result by calculating the gcd of result and the next number.
        return result
    
    #calculate the min deletions needed in nums to make the smallest number divide all numbers in numsDivide.
    def minDeletionsToDivide(self, nums, numsDivide):
        #compute the gcd of numsDivide using find_gcd
        gcd_value = self.find_gcd(numsDivide)
        
        #sort nums in ascending order
        nums.sort()
        
        #check for the smallest element in nums that divides the gcd
        for i, num in enumerate(nums):
            if gcd_value % num == 0: #check if current number divides the gcd
                return i  #returns the index of the number in the sorted nums. Number of deletions is the index of this element
        
        #if no such element is found, return -1
        return -1


nums = [2, 3, 2, 4, 3]
numsDivide = [9, 6, 9, 3, 15]
solution = Solution()
result = solution.minDeletionsToDivide(nums, numsDivide)
print(result)
```

```python
'''
QUESTION 48 - (135) - Candy


There are n children standing in a line. Each child is assigned a rating value given in the integer array ratings.
You are giving candies to these children subjected to the following requirements:

Each child must have at least one candy.
Children with a higher rating get more candies than their neighbors.

Return the minimum number of candies you need to have to distribute the candies to the children.


LOGIC:

we can start by initializing the candies array where each child gets 1 candy
candies = [1] * n

Left to Right Pass:

We iterate over the ratings array from left to right. If the current child's rating 
is greater than the previous child's rating, 
we increment the current child's candy count by 1 more than the previous child's candy count. 
This ensures that children with higher ratings than their neighbors get more candies.

We then iterate over the ratings array from right to left. 
If the current child's rating is greater than the next child's rating, 
and their candy count is not greater than the next child's candy count, 
we update the current child's candy count to ensure that they have more candies than the next child. 
The max() function ensures we don't reduce the candy count if it was already increased in the first pass.

The final result is the sum of all elements in the candies array, which gives the total number of candies required.





'''


def candy(ratings):
    n = len(ratings)
    
    #if there are no ratings (no children), no candy is needed
    if n == 0:
        return 0
    
    #initialize candies array where each child gets 1 candy initially.
    candies = [1] * n
    
    #traverse from left to right
    for i in range(1, n):
        if ratings[i] > ratings[i - 1]:
            candies[i] = candies[i - 1] + 1
    
    #traverse from right to left
    for i in range(n - 2, -1, -1):
        if ratings[i] > ratings[i + 1]:
            candies[i] = max(candies[i], candies[i + 1] + 1)
    
    #the answer is the sum of the candies array.
    return sum(candies)


ratings1 = [1, 0, 2]
ratings2 = [1, 2, 2]
print(candy(ratings1))
print(candy(ratings2))
```

```python
'''
QUESTION 49 - (224) - Basic Calculator

Given a string s representing a valid expression, implement a basic calculator to evaluate it, and return the result of the evaluation.
Note: You are not allowed to use any built-in function which evaluates strings as mathematical expressions, such as eval().


LOGIC:

building number process in the for loop
s= "123"
first 1: current_num = 1
then 2: current_num = 1 * 10 + 2 = 12
finally 3: current_num = 12 * 10 + 3 = 123

operation handling process
if sign = 1 and current_num = 5, add 5 to the result
if sign = -1 and current_num = 5, subtract 5 from the result.

parantheses handling process
push the current result and current sign into the stack
for e.g., (1 + 3)

stack = [0,1], then reset result = 0, sign = 1, to freshly calculate inside the paranthesis
next char is '1', so current_num = 0*10 + 1 = 1
next char is '+', sign stays as 1, apply sign and current_num to the result
result = 0 + 1 * 1 = 1
reset current_num and update the sign = 1 cause its a + case.
process the next char '3' the same way as '1' was processed, current_num = 3

processing of the closing bracket ')' -
apply the sign and current_num to the result
result = 1+1*3 = 4

pop the sign from the stack and multiply it with the result = 4*1 = 4
pop the previous result result from the stakc and add to result = 4 + 0 = 4

'''


def calculate(s: str) -> int:
    stack = []  #stack to hold parenthesis values
    current_num = 0  #to store the current number we're processing
    sign = 1  #tracks whether the current number should be added +1, or subtracted -1 to/from the result
    result = 0  #final result
    
    for i, char in enumerate(s): #loops through every character(char) in the string s[i].
        if char.isdigit(): #if the char is a digit
            # Build the current number and handle multiple digits
            current_num = current_num * 10 + int(char)
        
        #if the character is an operator, or open '(' or if we reached the end of the string
        if char in '+-(' or i == len(s) - 1:
            if char in '+-': #if char is + or - then handle the result accordingly.
                result += sign * current_num  #apply the previous operation
                current_num = 0  #reset the current number for the next number.
                sign = 1 if char == '+' else -1  #update sign based on operator
            
            elif char == '(': #handling open paranthesis case
                #push the result and sign into stack
                stack.append(result)
                stack.append(sign)
                result = 0  #reset result for the new subexpression
                sign = 1  #reset sign for the subexpression
            
            elif char == ')': #handling close paranthesis case
                result += sign * current_num  #apply the last operation for current number and finish the current calculation inside the paranthesis.
                current_num = 0  #reset current number
                result *= stack.pop()  #apply the sign before the parenthesis
                result += stack.pop()  #add the previous result before the parenthesis
            
    #the last number in the string should be added to result
    result += sign * current_num
    
    return result

s0 = "5 - 1"
s1 = "1 + 1"
s2 = " 2-1 + 2 "
s3 = "(1+(4+5+2)-3)+(6+8)"

print(calculate(s0))
print(calculate(s1))
print(calculate(s2))
print(calculate(s3))
```
